```python
import os
import pandas as pd

folder_path = 'xkom_do_scalenia'

all_dataframes = []

for file_name in sorted(os.listdir(folder_path)):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        print(f"Ładowanie pliku: {file_name}")

        df = pd.read_csv(file_path)
        all_dataframes.append(df)

full_dataframe = pd.concat(all_dataframes, ignore_index=True)

output_file = os.path.join(folder_path, 'produkty_xkom_szczegoly_full.csv')
full_dataframe.to_csv(output_file, index=False)

print(f"Scalono wszystkie pliki. Wynik zapisany jako: {output_file}")
```

```python
df = pd.read_csv('produkty_xkom_szczegoly_full.csv')
df
```

```python
df = df[~df['Cena'].str.contains('Brak ceny', na=False)]
df = df.dropna(subset=['Cena'])

df['Cena'] = (
    df['Cena']
    .str.replace('zł', '', regex=False)
    .str.replace(',', '')
    .str.replace(' ', '')
    .astype(int)
)

df['Cena'] = df.apply(
    lambda row: row['Cena'] / 100,
    axis=1
)

df
```

```python
df = df.rename(columns={'Pamięć': 'RAM'})

df['RAM'] = df['RAM'].str.extract(r'^(\d+)', expand=False)

df['RAM'] = df['RAM'].astype(int)

df.head()
```

```python
df['Dysk'] = df['Tytuł'].str.extract(r'GB/([^/]+)/', expand=False)
df
```

```python
def przetworz_dysk(wartosc):
    if pd.isna(wartosc):
        return 0
    if '+' in wartosc:
        czesci = wartosc.split('+')
        suma = 0
        for czesc in czesci:
            suma += przetworz_dysk(czesc.strip())
        return suma
    if 'TB' in wartosc:
        try:
            return int(wartosc.replace('TB', '').strip()) * 1024
        except ValueError:
            return 0
    if 'GB' in wartosc:
        try:
            return int(wartosc.replace('GB', '').strip())
        except ValueError:
            return 0
    try:
        return int(wartosc.strip())
    except ValueError:
        return 0


df['Dysk GB'] = df['Dysk'].apply(przetworz_dysk)
df = df.rename(columns={'Dysk GB': 'Dysk SSD'})
```

```python
def wyodrebnij_system(tytul):
    tytul = tytul.lower()  # Ujednolicamy do małych liter dla większej elastyczności
    if 'mac os' in tytul or 'macos' in tytul:
        return 'macOS Sonoma'
    elif 'win11p' in tytul or 'w11p' in tytul:
        return 'Windows 11 Professional'
    elif 'win11' in tytul or 'w11' in tytul:
        return 'Windows 11 Home'
    elif 'chromeos' in tytul or 'chromebook' in tytul:
        return 'Chrome OS'
    elif 'freedos' in tytul:
        return 'FreeDOS'
    else:
        return 'Brak'


df['System operacyjny'] = df['Tytuł'].apply(wyodrebnij_system)
```

```python
def extract_processor_brand(processor_name):
    if pd.isna(processor_name):
        return 'Brak'
    if 'Intel' in processor_name:
        return 'Intel'
    elif 'AMD' in processor_name:
        return 'AMD'
    elif 'Apple' in processor_name:
        return 'Apple'
    elif 'MediaTek' in processor_name:
        return 'MediaTek'
    elif 'Qualcomm' in processor_name:
        return 'Qualcomm'
    else:
        return 'Inne'

df['Firma procesora'] = df['Procesor'].apply(extract_processor_brand)
df
```
