```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

ds = Service()
driver = webdriver.Chrome(service=ds)

products_data = []

# range(1,6)
page = 2
url = f"https://www.mediaexpert.pl/komputery-i-tablety/laptopy-i-ultrabooki/laptopy?page={page}"
driver.get(url)

try:
    accept_cookies_button = driver.find_element(By.ID, "onetrust-accept-btn-handler")
    accept_cookies_button.click()
    print("Kliknięto przycisk akceptacji ciasteczek")
except:
    print("Przycisk akceptacji ciasteczek nie został znaleziony")

products = driver.find_elements(By.XPATH, "//span/div[contains(@class, 'offer-box')]")

for product in products:
    try:
        title = product.find_element(By.XPATH, ".//h2/a[contains(@class, 'is-animate ui-link')]").text
    except:
        title = "Brak danych"

    try:
        price = product.find_element(By.XPATH, ".//div[contains(@class, 'main-price')]").text
    except:
        price = "Brak ceny"

    try:
        table = product.find_element(By.XPATH, ".//table[contains(@class, 'list attributes')]")
        rows = table.find_elements(By.XPATH, "./tbody/tr")

        details = {}
        for row in rows:
            try:
                column_name = row.find_element(By.XPATH, "./th/span").text.strip()
                value = row.find_element(By.XPATH, "./td/span").text.strip()
                details[column_name] = value
            except Exception as e:
                print(f"Błąd w przetwarzaniu szczegółów: {e}")
    except:
        details = {}

    product_data = {
        "Tytuł": title,
        "Cena": price,
        "Sklep": "ediaexpert"
    }
    product_data.update(details)

    products_data.append(product_data)

print(f"--- Zakończono stronę {page} ---")

import pandas as pd

df = pd.DataFrame(products_data)
df.to_csv(f"produkty_mediaexpert_szczegoly_{page}.csv", index=False)
print("Dane zapisano do pliku produkty_mediaexpert_szczegoly.csv")

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd

for page in range(7, 75):
    ds = Service()
    driver = webdriver.Chrome(service=ds)

    products_data_xkom = []
    url = f"https://www.x-kom.pl/g-2/c/159-laptopy-notebooki-ultrabooki.html?page={page}"
    driver.get(url)

    time.sleep(5)

    products = driver.find_elements(By.XPATH,
                                    "//div[contains(@class, 'BoxFlex__Box-sc-bfe0bca9-0 iHPBSK parts__ProductListCol-sc-a93b49f9-1 coJOdn')]")

    for product in products:
        title = "Brak danych"
        price = "Brak danych"
        details = {}

        try:
            title = product.find_element(By.XPATH,
                                         ".//h3[contains(@class, 'parts__Title-sc-1d28d-0 fKUIM parts__Title-sc-6e280ffa-9 hqUuGB')]/span").text
        except Exception as e:
            print("Nie udało się pobrać tytułu dla produktu:", e)

        try:
            price = product.find_element(By.XPATH,
                                         ".//span[contains(@class, 'sc-jnqLxu cjLwnY parts__Price-sc-6e255ce0-0 jWNFsL')]").text
        except Exception as e:
            print("Nie udało się pobrać ceny dla produktu:", e)

        try:
            features = product.find_elements(By.XPATH, ".//li[contains(@class, 'parts__ListItem-sc-27c8b4c9-2 eDfVv')]")
            for feature in features:
                try:
                    text = feature.text.strip()
                    if ":" in text:
                        column_name, column_value = text.split(":", 1)  # Rozdzielamy tylko raz
                        details[column_name.strip()] = column_value.strip()
                except Exception as e:
                    print(f"Błąd przy przetwarzaniu elementu <li>: {e}")
        except Exception as e:
            print("Nie udało się pobrać szczegółów dla produktu:", e)

        product_data = {
            "Tytuł": title,
            "Cena": price,
            "Sklep": "xkom"
        }
        product_data.update(details)
        print(f"--- title: {title} ---")
        print(f"--- price: {price} ---")

        products_data_xkom.append(product_data)

    print(f"--- Zakończono stronę {page} ---")

    df = pd.DataFrame(products_data_xkom)
    df.to_csv(f"produkty_xkom_szczegoly_{page}.csv", index=False)
    print(f"Dane zapisano do pliku produkty_xkom_szczegoly_{page}.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd

ds = Service()
driver = webdriver.Chrome(service=ds)

products_data_rtv = []

for page in range(1, 66):
    url = f"https://www.euro.com.pl/laptopy-i-netbooki,strona-{page}.bhtml?link=old-nav"
    driver.get(url)

    time.sleep(3)

    product_containers = driver.find_elements(By.XPATH,
                                              "//div[contains(@class, 'product-medium-box ng-star-inserted')]")

    for product in product_containers:
        try:
            title_element = product.find_element(By.XPATH, ".//a[contains(@class, 'product-medium-box-intro__link')]")
            title = title_element.text

            link = title_element.get_attribute("href")

            print(f"Tytuł: {title}")
        except Exception as e:
            print("Błąd podczas pobierania danych produktu:", e)
            title = "Brak tytułu"
            link = "Brak linku"

        try:
            price = product.find_element(By.XPATH,
                                         ".//div[contains(@class, 'parted-price ng-star-inserted')]/span[contains(@class, 'parted-price-total')]").text
            print(f"Cena: {price}")
        except Exception as e:
            print("Błąd podczas pobierania ceny produktu:", e)
            price = "Brak ceny"

        technical_details = {}
        try:
            technical_data_list = product.find_elements(By.XPATH, ".//ul[contains(@class, 'technical-data__list')]/li")
            for item in technical_data_list:
                try:
                    column_title = item.find_element(By.XPATH, ".//span[contains(@class, 'technical-data__list-item-name')]").text

                    column_value = item.find_element(By.XPATH, ".//span[contains(@class, 'technical-data-item-text')]").text

                    technical_details[column_title] = column_value

                    print(f"{column_title}: {column_value}")
                except Exception as e:
                    print(f"Błąd w przetwarzaniu elementu: {e}")
        except Exception as e:
            print(f"Błąd w odnalezieniu sekcji danych technicznych: {e}")

        product_data = {"Tytuł": title, "Cena": price, "Link": link}
        product_data.update(technical_details)

        products_data_rtv.append(product_data)

df = pd.DataFrame(products_data_rtv)
df.to_csv("produkty_rtv_agd_full.csv", index=False)
print("Dane zapisano do pliku produkty_rtv_agd_full.csv")

```
