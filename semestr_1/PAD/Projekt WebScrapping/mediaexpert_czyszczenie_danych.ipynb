```python
import pandas as pd
import numpy as np

df = pd.read_csv('produkty_mediaexpert_szczegoly_full.csv')
df
```

```python
df = df[~df['Cena'].str.contains('Brak ceny', na=False)]
df = df.dropna(subset=['Cena'])

df['Cena'] = (
    df['Cena']
    .str.replace('\n', '', regex=True)
    .str.replace('zł', '', regex=False)
    .str.replace('\u202f', '', regex=False)
    .str.replace('n', '', regex=False)
    .astype(int)
)

df['Cena'] = df.apply(
    lambda row: row['Cena'] /100,
    axis=1
)

df

```

```python
df
```

```python
df = df.rename(columns={'RAM:': 'RAM'})
df['RAM'] = df['RAM'].str.extract(r'(\d+)\s*GB', expand=False).replace('', float('nan'))
df.head()
```

```python
df = df.rename(columns={'Dysk SSD:': 'Dysk SSD'})
df[['Dysk_Wartosc', 'Dysk_Jednostka']] = df['Dysk SSD'].str.extract(r'(\d+)\s*(GB|TB)', expand=True)
df['Dysk_Wartosc'] = df.apply(
    lambda row: row['Dysk_Wartosc'] * 1024 if row['Dysk_Jednostka'] == 'TB' else row['Dysk_Wartosc'],
    axis=1
)
df['Dysk SSD'] = df['Dysk_Wartosc']

df = df.drop(columns=['Dysk_Wartosc', 'Dysk_Jednostka'])

df
```

```python
df = df.rename(columns={'System operacyjny:': 'System operacyjny'})
df['System operacyjny'].value_counts()
```

```python
df['System operacyjny'].value_counts()
def wyodrebnij_system(tytul):
    if pd.isna(tytul):
        return 'Brak'
    tytul = tytul.lower()
    if 'macos' in tytul:
        return 'macOS Sonoma'
    elif 'windows 11 home' in tytul or 'windows 11 s' in tytul:
        return 'Windows 11 Home'
    elif 'windows 11 professional' in tytul:
        return 'Windows 11 Professional'
    elif 'windows 10' in tytul:
        return 'Windows 10'
    elif 'chrome os' in tytul or 'chromebook' in tytul:
        return 'Chrome OS'
    elif 'freedos' in tytul:
        return 'FreeDOS'
    elif 'linux' in tytul:
        return 'Linux'
    else:
        return 'Brak'

df['System operacyjny'] = df['System operacyjny'].apply(wyodrebnij_system)
```

```python
df['System operacyjny'].value_counts()
```

```python
df = df.rename(columns={'Procesor:': 'Procesor'})
df['Procesor'].value_counts()
```

```python
def extract_processor_brand(processor_name):
    if 'Intel' in processor_name:
        return 'Intel'
    elif 'AMD' in processor_name:
        return 'AMD'
    elif 'Apple' in processor_name:
        return 'Apple'
    elif 'MediaTek' in processor_name:
        return 'MediaTek'
    elif 'Qualcomm' in processor_name:
        return 'Qualcomm'
    else:
        return 'Inne'

df['Firma procesora'] = df['Procesor'].apply(extract_processor_brand)
df
```

```python

df['Firma procesora'].value_counts()
df['Sklep'] = 'MediaExpert'
df['Firma procesora'].value_counts()
```

```python
df = df.rename(columns={'Ekran:': 'Ekran'})
df['Ekran'] = df['Ekran'].str.extract(r'(\d+(?:\.\d+)?)', expand=False)
```

```python
df.to_csv('produkty_mediaexpert_szczegoly_full_cleaned.csv', index=False, encoding='utf-8')
```
